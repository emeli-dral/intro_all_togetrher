```python
import pandas as pd
```

```python
data = pd.read_csv('german_data.tsv', header=None, sep = '[\s]*', engine='python')
```

```python
data.columns = map( lambda x  :  'A' + str(x), data.columns)
```

## Model traning

```python
from sklearn import cross_validation, metrics, linear_model
```

```python
data.ix[:, :-1].head()
```

```python
data.shape
```

```python
?cross_validation.train_test_split
```

```python
X_train, X_test, y_train, y_test = cross_validation.train_test_split(data.ix[:, :-1], data.A24, test_size = 0.5)
```

```python
y_train.value_counts()
```

```python
model = linear_model.LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predictions = model.predict(X_test)
```

```python
y_predictions
```

```python
metrics.accuracy_score(y_test, y_predictions)
```

```python
[1]*len(y_test)
```

```python
metrics.accuracy_score(y_test, [1]*len(y_test))
```

```python
print metrics.classification_report(y_test, y_predictions)
```

```python
print metrics.classification_report(y_train, model.predict(X_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
neighbors_model = KNeighborsClassifier(n_neighbors = 15)
```

```python
neighbors_model.fit(X_train, y_train)
```

```python
metrics.accuracy_score(y_test, neighbors_model.predict(X_test))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.transform(X_test)
```

```python
neighbors_model.fit(X_train_scaled, y_train)
```

```python
metrics.accuracy_score(y_test, neighbors_model.predict(X_test_scaled))
```

```python
neighbors_model = KNeighborsClassifier(n_neighbors = 5)
```

```python
neighbors_model.fit(X_train_scaled, y_train)
metrics.accuracy_score(y_test, neighbors_model.predict(X_test_scaled)
```

```python
n_neighbors = [1, 2, 5, 10, 15, 20, 25, 30 , 50]
```

```python
%pylab inline
```

```python
plot(n_neighbors)
```

```python
results = []
for k in n_neighbors:
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(X_train_scaled, y_train)
    results.append(
        metrics.accuracy_score(
            model.predict(X_test), y_test)
    )
```

```python
plot(n_neighbors, results)
```

```python
from sklearn.grid_search import GridSearchCV
```

```python
grid = {'n_neighbors': [1, 2, 5, 10, 15, 20, 25, 30 , 50]}
```

```python
grid
```

```python
?GridSearchCV
```

```python
model = KNeighborsClassifier()
```

```python
grid_search = GridSearchCV(model, grid, scoring='accuracy', cv=5)
```

```python
grid_search.fit(X_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_.predict(X_test_scaled)
```
